import sympy as sp

# Define variables
x, y = sp.symbols('x y')

# --- Task 1: First-Order Ordinary Differential Equation y' = y ---
# Define the differential equation: y' = y
lhs = sp.Function('y')(x).diff(x)
rhs = sp.Function('y')(x)

# Solve the differential equation using separation of variables
solution1 = sp.dsolve(lhs - rhs)
print("Solution to y' = y:", solution1)


# --- Task 2: First-Order Ordinary Differential Equation y' = (1/2) * y ---
# Define the differential equation: y' = (1/2) * y
lhs2 = sp.Function('y')(x).diff(x)
rhs2 = (1/2) * sp.Function('y')(x)

# Solve the differential equation using separation of variables
solution2 = sp.dsolve(lhs2 - rhs2)
print("\nSolution to y' = (1/2) * y:", solution2)


# --- Task 3: Second-Order Ordinary Differential Equation y'' + y' = 0 with boundary conditions y(0) = 2, y'(0) = -1 ---
# Define the second-order differential equation: y'' + y' = 0
y2 = sp.Function('y')(x)
lhs3 = y2.diff(x, 2) + y2.diff(x)

# Solve the differential equation with initial conditions y(0) = 2, y'(0) = -1
solution3 = sp.dsolve(lhs3, ics={y2.subs(x, 0): 2, y2.diff(x).subs(x, 0): -1})
print("\nSolution to y'' + y' = 0 with y(0) = 2, y'(0) = -1:", solution3)


# --- Task 4: Second-Order Ordinary Differential Equation y'' - y = 0 with boundary conditions y(0) = 2, y'(0) = 0 ---
# Define the second-order differential equation: y'' - y = 0
lhs4 = y2.diff(x, 2) - y2

# Solve the differential equation with initial conditions y(0) = 2, y'(0) = 0
solution4 = sp.dsolve(lhs4, ics={y2.subs(x, 0): 2, y2.diff(x).subs(x, 0): 0})
print("\nSolution to y'' - y = 0 with y(0) = 2, y'(0) = 0:", solution4)

ω = sp.symbols('ω')

# Define the differential equation: y'' = -ω²y
lhs5 = y2.diff(x, 2) + ω**2 * y2

# Solve the differential equation
solution5 = sp.dsolve(lhs5)
print("\nSolution to y'' = -ω²y:", solution5)
