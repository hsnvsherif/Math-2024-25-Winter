## Exercises for Students

Solve the following systems of equations using row operations and matrix methods (similar to the example above):

### 1. System 1:
\[
3x - 2y = 5, \quad 2x + 3y = 7
\]

### 2. System 2:
\[
2x - 3y = 10, \quad 4x + 5y = 20
\]

### 3. System 3:
\[
2x - y + z = 3, \quad x + 2y - z = 1, \quad 3x - y + 2z = 11
\]

### 4. System 4:
\[
2x - 3y + 4z + 2t = 2, \quad 3x + 2y - 5z + 3t = 3, \quad 4x - 3y + 2z - 5t = 4, \quad 5x + 4y - 3z + 2t = 5
\]


```python
import numpy as np

# Define matrices (coefficients and constants)
A1 = np.array([[3, -2], [2, 3]])
B1 = np.array([5, 7])

A2 = np.array([[2, -3], [4, 5]])
B2 = np.array([10, 20])

A3 = np.array([[2, -1, 1], [1, 2, -1], [3, -1, 2]])
B3 = np.array([3, 1, 11])

A4 = np.array([[2, -3, 4, 2], [3, 2, -5, 3], [4, -3, 2, -5], [5, 4, -3, 2]])
B4 = np.array([2, 3, 4, 5])

# Solve using numpy.linalg.solve
x1, y1 = np.linalg.solve(A1, B1)
x2, y2 = np.linalg.solve(A2, B2)
x3, y3, z3 = np.linalg.solve(A3, B3)
x4, y4, z4, t4 = np.linalg.solve(A4, B4)

# Print the solutions
print("Solution for System 1: 3x - 2y = 5, 2x + 3y = 7")
print(f"x = {x1:.2f}, y = {y1:.2f}\n")

print("Solution for System 2: 2x - 3y = 10, 4x + 5y = 20")
print(f"x = {x2:.2f}, y = {y2:.2f}\n")

print("Solution for System 3: 2x - y + z = 3, x + 2y - z = 1, 3x - y + 2z = 11")
print(f"x = {x3:.2f}, y = {y3:.2f}, z = {z3:.2f}\n")

print("Solution for System 4: 2x - 3y + 4z + 2t = 2, 3x + 2y - 5z + 3t = 3, 4x - 3y + 2z - 5t = 4, 5x + 4y - 3z + 2t = 5")
print(f"x = {x4:.2f}, y = {y4:.2f}, z = {z4:.2f}, t = {t4:.2f}\n")
```
