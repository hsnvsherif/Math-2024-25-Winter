## Exercises for Students

### System 1:
Solve the system of equations:
$$
\left\{
\begin{array}{l}
x + 2y - 2z = 4 \\
2x + y + z = 0 \\
3x + 2y + z = 1
\end{array}
\right.
$$

### System 2:
Solve the system of equations:
$$
\left\{
\begin{array}{l}
x + y + z - t = 2 \\
2x + y + z = 3 \\
-x + z - t = 0 \\
3x + 2y - z + 2t = -1
\end{array}
\right.
$$

### System 3:
Solve the system of equations:
$$
\left\{
\begin{array}{l}
x + y - z - t = 0 \\
2x + 3y - 2z + t = 4 \\
3x + 5z = 0 \\
-x + y - 3z + 2t = 3
\end{array}
\right.
$$

---

### Python Code to Solve the Systems

```python
import sympy as sp

# Define symbols
x, y, z, t = sp.symbols('x y z t')

# System 1
eq1_1 = sp.Eq(x + 2*y - 2*z, 4)
eq1_2 = sp.Eq(2*x + y + z, 0)
eq1_3 = sp.Eq(3*x + 2*y + z, 1)
solution_1 = sp.solve([eq1_1, eq1_2, eq1_3], (x, y, z))

# System 2
eq2_1 = sp.Eq(x + y + z - t, 2)
eq2_2 = sp.Eq(2*x + y + z, 3)
eq2_3 = sp.Eq(-x + z - t, 0)
eq2_4 = sp.Eq(3*x + 2*y - z + 2*t, -1)
solution_2 = sp.solve([eq2_1, eq2_2, eq2_3, eq2_4], (x, y, z, t))

# System 3
eq3_1 = sp.Eq(x + y - z - t, 0)
eq3_2 = sp.Eq(2*x + 3*y - 2*z + t, 4)
eq3_3 = sp.Eq(3*x + 5*z, 0)
eq3_4 = sp.Eq(-x + y - 3*z + 2*t, 3)
solution_3 = sp.solve([eq3_1, eq3_2, eq3_3, eq3_4], (x, y, z, t))

# Display the solutions
print("Solution to System 1:", solution_1)
print("Solution to System 2:", solution_2)
print("Solution to System 3:", solution_3)
