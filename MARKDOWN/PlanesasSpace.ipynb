AC = C - A

# Find the normal vector to the plane (cross product of AB and AC)

\text{Given:} \\
\text{Plane 1: } x + y + z = 1 \quad \text{with normal vector } \mathbf{n_1} = [1, 1, 1] \\
\text{Plane 2: } x = 0 \quad \text{with normal vector } \mathbf{n_2} = [1, 0, 0] \\
\text{The angle between the planes:} \\
\cos(\theta) = \frac{\mathbf{n_1} \cdot \mathbf{n_2}}{|\mathbf{n_1}| |\mathbf{n_2}|} \\
\mathbf{n_1} \cdot \mathbf{n_2} = 1 \times 1 + 1 \times 0 + 1 \times 0 = 1 \\
|\mathbf{n_1}| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}, \quad |\mathbf{n_2}| = \sqrt{1^2} = 1 \\
\cos(\theta) = \frac{1}{\sqrt{3}} \\
\theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right)
