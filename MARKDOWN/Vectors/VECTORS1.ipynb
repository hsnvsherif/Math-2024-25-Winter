A. By what number should vector $\mathbf{a} = [3, 4]$ be multiplied so that its length is equal to 1?

To make the length of vector $\mathbf{a} = [3, 4]$ equal to 1, we need to find the scaling factor $k$ such that the magnitude of the scaled vector equals 1.

The length (magnitude) of a vector $\mathbf{a} = [a_x, a_y]$ is given by:

$$
\|\mathbf{a}\| = \sqrt{a_x^2 + a_y^2}
$$

For $\mathbf{a} = [3, 4]$, the magnitude is:

$$
\|\mathbf{a}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

To scale the vector $\mathbf{a}$ such that its length is 1, we multiply it by a factor $k$:

$$
\|\mathbf{a'}\| = k \|\mathbf{a}\| = 1
$$

Thus, we need:

$$
k \cdot 5 = 1
$$

Solving for $k$:

$$
k = \frac{1}{5}
$$

So, the vector $\mathbf{a}$ should be multiplied by $k = \frac{1}{5}$ to make its length equal to 1.

B. Calculate the length of vector $\mathbf{b} = [1, 1]$ and find the unit vector of this vector.

To calculate the length of vector $\mathbf{b} = [1, 1]$ and find its unit vector, we follow these steps:

### 1. Calculate the Length (Magnitude) of the Vector

The length (magnitude) of a vector $\mathbf{b} = [b_x, b_y]$ is given by:

$$
\|\mathbf{b}\| = \sqrt{b_x^2 + b_y^2}
$$

For $\mathbf{b} = [1, 1]$, the magnitude is:

$$
\|\mathbf{b}\| = \sqrt{1^2 + 1^2} = \sqrt{1 + 1} = \sqrt{2}
$$

So, the length of vector $\mathbf{b}$ is $\|\mathbf{b}\| = \sqrt{2}$.

### 2. Find the Unit Vector

The unit vector $\hat{\mathbf{b}}$ in the direction of $\mathbf{b}$ is obtained by dividing the vector $\mathbf{b}$ by its magnitude:

$$
\hat{\mathbf{b}} = \frac{\mathbf{b}}{\|\mathbf{b}\|}
$$

Thus, the unit vector is:

$$
\hat{\mathbf{b}} = \frac{[1, 1]}{\sqrt{2}} = \left[\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right]
$$

Therefore, the unit vector in the direction of $\mathbf{b}$ is $\hat{\mathbf{b}} = \left[\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right]$.

D. Calculate the length of vector $\mathbf{c} = [1, 2, 3]$ and find the unit vector of this vector.

### D. Calculate the Length of Vector $\mathbf{c} = [1, 2, 3]$ and Find the Unit Vector of This Vector

#### 1. Calculate the Length (Magnitude) of the Vector

The length (magnitude) of a vector $\mathbf{c} = [c_x, c_y, c_z]$ in three-dimensional space is given by:

$$
\|\mathbf{c}\| = \sqrt{c_x^2 + c_y^2 + c_z^2}
$$

For $\mathbf{c} = [1, 2, 3]$, the magnitude is:

$$
\|\mathbf{c}\| = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{1 + 4 + 9} = \sqrt{14}
$$

So, the length of vector $\mathbf{c}$ is $\|\mathbf{c}\| = \sqrt{14}$.

#### 2. Find the Unit Vector

The unit vector $\hat{\mathbf{c}}$ in the direction of $\mathbf{c}$ is obtained by dividing the vector $\mathbf{c}$ by its magnitude:

$$
\hat{\mathbf{c}} = \frac{\mathbf{c}}{\|\mathbf{c}\|}
$$

Thus, the unit vector is:

$$
\hat{\mathbf{c}} = \frac{[1, 2, 3]}{\sqrt{14}} = \left[\frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}}\right]
$$

Therefore, the unit vector in the direction of $\mathbf{c}$ is:

$$
\hat{\mathbf{c}} = \left[\frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}}\right]
$$

E. Find the Cartesian coordinates of vector $\mathbf{v} = [2, 3, 4]$ in the basis $\{\mathbf{b_1} = [1, 0, 1], \mathbf{b_2} = [0, 1, 0], \mathbf{b_3} = [1, 0, -1]\}$.


### E. Find the Cartesian Coordinates of Vector $\mathbf{v} = [2, 3, 4]$ in the Basis $\{\mathbf{b_1} = [1, 0, 1], \mathbf{b_2} = [0, 1, 0], \mathbf{b_3} = [1, 0, -1]\}$

We express the vector $\mathbf{v}$ as a linear combination of the new basis vectors:

$$
\mathbf{v} = c_1 \mathbf{b_1} + c_2 \mathbf{b_2} + c_3 \mathbf{b_3}
$$

Substituting the given values:

$$
[2, 3, 4] = c_1 [1, 0, 1] + c_2 [0, 1, 0] + c_3 [1, 0, -1]
$$

This gives the system of equations:

1. $c_1 + c_3 = 2$
2. $c_2 = 3$
3. $c_1 - c_3 = 4$

From the second equation, we know that:

$$
c_2 = 3
$$

Substitute $c_2 = 3$ into the first and third equations:

1. $c_1 + c_3 = 2$
2. $c_1 - c_3 = 4$

Adding these two equations to eliminate $c_3$:

$$
(c_1 + c_3) + (c_1 - c_3) = 2 + 4
$$

$$
2c_1 = 6 \quad \Rightarrow \quad c_1 = 3
$$

Substitute $c_1 = 3$ into $c_1 + c_3 = 2$:

$$
3 + c_3 = 2 \quad \Rightarrow \quad c_3 = -1
$$

Thus, the Cartesian coordinates of $\mathbf{v}$ in the new basis are:

$$
c_1 = 3, \quad c_2 = 3, \quad c_3 = -1
$$

Therefore, $\mathbf{v}$ can be written as:

$$
\mathbf{v} = 3 \mathbf{b_1} + 3 \mathbf{b_2} - 1 \mathbf{b_3}
$$
