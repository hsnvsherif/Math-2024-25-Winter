\text{Given the spheres:} \\
S_1: x^2 + y^2 + z^2 = 1 \\
S_2: x^2 + y^2 + z^2 = 2 \\
\text{Since the radii of the spheres are different, the spheres do not intersect.} \\
\text{Radius of } S_1 = 1, \quad \text{Radius of } S_2 = \sqrt{2}. \\
\text{Thus, the spheres do not have any common points.}

8:3

---

\text{Given the two spheres:} \\
x^2 + y^2 + z^2 = 1 \quad \text{(Sphere 1)} \\
(x - 1)^2 + y^2 + z^2 = 1 \quad \text{(Sphere 2)} \\
\text{Subtracting the second equation from the first:} \\
x^2 + y^2 + z^2 - \left[ (x - 1)^2 + y^2 + z^2 \right] = 0 \\
x^2 + y^2 + z^2 - \left( (x - 1)^2 + y^2 + z^2 \right) = 0 \\
x^2 + y^2 + z^2 - x^2 + 2x - 1 - y^2 - z^2 = 0 \\
2x - 1 = 0 \\
x = \frac{1}{2} \\
\text{Substitute } x = \frac{1}{2} \text{ into the first sphere equation:} \\
\left( \frac{1}{2} \right)^2 + y^2 + z^2 = 1 \\
\frac{1}{4} + y^2 + z^2 = 1 \\
y^2 + z^2 = 1 - \frac{1}{4} \\
y^2 + z^2 = \frac{3}{4} \\
\text{Thus, the equation of the curve is:} \\
x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}
