### 1. System of Equations: 
$$
3x - 2y = 5, \quad 2x + 3y = 7
$$

#### Solution:

1. Write the augmented matrix:

$$
\begin{bmatrix}
3 & -2 & | & 5 \\
2 & 3 & | & 7
\end{bmatrix}
$$

2. Perform row operations:

- Divide row 1 by 3 to normalize the first element of row 1:

$$ R_1 \rightarrow \frac{1}{3} R_1 $$

$$
\begin{bmatrix}
1 & -\frac{2}{3} & | & \frac{5}{3} \\
2 & 3 & | & 7
\end{bmatrix}
$$

- Subtract 2 times row 1 from row 2 to eliminate the first element of row 2:

$$ R_2 \rightarrow R_2 - 2R_1 $$

$$
\begin{bmatrix}
1 & -\frac{2}{3} & | & \frac{5}{3} \\
0 & \frac{11}{3} & | & \frac{17}{3}
\end{bmatrix}
$$

- Multiply row 2 by \( \frac{3}{11} \) to normalize the second pivot element:

$$ R_2 \rightarrow \frac{3}{11} R_2 $$

$$
\begin{bmatrix}
1 & -\frac{2}{3} & | & \frac{5}{3} \\
0 & 1 & | & \frac{17}{11}
\end{bmatrix}
$$

- Add \( \frac{2}{3} \) times row 2 to row 1 to eliminate the second element of row 1:

$$ R_1 \rightarrow R_1 + \frac{2}{3}R_2 $$

$$
\begin{bmatrix}
1 & 0 & | & \frac{39}{33} \\
0 & 1 & | & \frac{17}{11}
\end{bmatrix}
$$

3. The solution is:

$$
x = \frac{39}{33}, \quad y = \frac{17}{11}
$$

### 2. System of Equations: 
$$
2x - 3y = 10, \quad 4x + 5y = 20
$$

#### Solution:

1. Write the augmented matrix:

$$
\begin{bmatrix}
2 & -3 & | & 10 \\
4 & 5 & | & 20
\end{bmatrix}
$$

2. Perform row operations:

- Divide row 1 by 2:

$$ R_1 \rightarrow \frac{1}{2} R_1 $$

$$
\begin{bmatrix}
1 & -\frac{3}{2} & | & 5 \\
4 & 5 & | & 20
\end{bmatrix}
$$

- Subtract 4 times row 1 from row 2:

$$ R_2 \rightarrow R_2 - 4R_1 $$

$$
\begin{bmatrix}
1 & -\frac{3}{2} & | & 5 \\
0 & 13 & | & 0
\end{bmatrix}
$$

- Divide row 2 by 13 to normalize the second pivot:

$$ R_2 \rightarrow \frac{1}{13} R_2 $$

$$
\begin{bmatrix}
1 & -\frac{3}{2} & | & 5 \\
0 & 1 & | & 0
\end{bmatrix}
$$

- Add \( \frac{3}{2} \) times row 2 to row 1 to eliminate the second element of row 1:

$$ R_1 \rightarrow R_1 + \frac{3}{2}R_2 $$

$$
\begin{bmatrix}
1 & 0 & | & 5 \\
0 & 1 & | & 0
\end{bmatrix}
$$

3. The solution is:

$$
x = 5, \quad y = 0
$$

### 3. System of Equations: 
$$
2x - y + z = 3, \quad x + 2y - z = 1, \quad 3x - y + 2z = 11
$$

#### Solution:

1. Write the augmented matrix:

$$
\begin{bmatrix}
2 & -1 & 1 & | & 3 \\
1 & 2 & -1 & | & 1 \\
3 & -1 & 2 & | & 11
\end{bmatrix}
$$

2. Perform row operations:

- Swap row 1 and row 2 to make the pivot of row 1 1:

$$ R_1 \leftrightarrow R_2 $$

$$
\begin{bmatrix}
1 & 2 & -1 & | & 1 \\
2 & -1 & 1 & | & 3 \\
3 & -1 & 2 & | & 11
\end{bmatrix}
$$

- Subtract 2 times row 1 from row 2:

$$ R_2 \rightarrow R_2 - 2R_1 $$

$$
\begin{bmatrix}
1 & 2 & -1 & | & 1 \\
0 & -5 & 3 & | & 1 \\
3 & -1 & 2 & | & 11
\end{bmatrix}
$$

- Subtract 3 times row 1 from row 3:

$$ R_3 \rightarrow R_3 - 3R_1 $$

$$
\begin{bmatrix}
1 & 2 & -1 & | & 1 \\
0 & -5 & 3 & | & 1 \\
0 & -7 & 5 & | & 8
\end{bmatrix}
$$

- Divide row 2 by \( -5 \):

$$ R_2 \rightarrow \frac{1}{-5} R_2 $$

$$
\begin{bmatrix}
1 & 2 & -1 & | & 1 \\
0 & 1 & -\frac{3}{5} & | & -\frac{1}{5} \\
0 & -7 & 5 & | & 8
\end{bmatrix}
$$

- Add 7 times row 2 to row 3:

$$ R_3 \rightarrow R_3 + 7R_2 $$

$$
\begin{bmatrix}
1 & 2 & -1 & | & 1 \\
0 & 1 & -\frac{3}{5} & | & -\frac{1}{5} \\
0 & 0 & \frac{16}{5} & | & \frac{43}{5}
\end{bmatrix}
$$

- Multiply row 3 by \( \frac{5}{16} \):

$$ R_3 \rightarrow \frac{5}{16} R_3 $$

$$
\begin{bmatrix}
1 & 2 & -1 & | & 1 \\
0 & 1 & -\frac{3}{5} & | & -\frac{1}{5} \\
0 & 0 & 1 & | & \frac{43}{16}
\end{bmatrix}
$$

- Add \( \frac{3}{5} \) times row 3 to row 2 and 1 times row 3 to row 1:

$$ R_2 \rightarrow R_2 + \frac{3}{5} R_3 $$  
$$ R_1 \rightarrow R_1 + R_3 $$

$$
\begin{bmatrix}
1 & 2 & 0 & | & \frac{59}{16} \\
0 & 1 & 0 & | & \frac{7}{16} \\
0 & 0 & 1 & | & \frac{43}{16}
\end{bmatrix}
$$

3. The solution is:

$$
x = \frac{59}{16}, \quad y = \frac{7}{16}, \quad z = \frac{43}{16}
$$

### 4. System of Equations: 
$$
2x - 3y + 4z + 2t = 2, \quad 3x + 2y - 5z + 3t = 3, \quad 4x - 3y + 2z - 5t = 4, \quad 5x + 4y - 3z + 2t = 5
$$

#### Solution:

1. Write the augmented matrix:

$$
\begin{bmatrix}
2 & -3 & 4 & 2 & | & 2 \\
3 & 2 & -5 & 3 & | & 3 \\
4 & -3 & 2 & -5 & | & 4 \\
5 & 4 & -3 & 2 & | & 5
\end{bmatrix}
$$

2. Perform row operations to reduce the matrix and find the solution.
