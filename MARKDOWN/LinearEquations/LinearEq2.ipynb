## Exercises for Students

1. Solve the system of equations:
   $$
   \begin{cases}
   2x_1 - 3x_2 = 7 \\
   3x_1 + 5x_2 = 2
   \end{cases}
   $$

2. Solve the system of equations:
   $$
   \begin{cases}
   2x + y - z = 1 \\
   x - y + 2z = 4 \\
   3x - 2z = -1
   \end{cases}
   $$

3. Solve the system of equations:
   $$
   \begin{cases}
   x + y + z - t = 2 \\
   x - z + 2t = 6 \\
   2x - 3y + t = 4 \\
   3x + y + 3z - 4t = -2
   \end{cases}
   $$

4. Why can't the following system of equations be solved using Cramer's rule?
$$
\begin{cases}
x_1 + 2x_2 + 3x_3 = 3 \\
4x_1 + 5x_2 + 6x_3 = 2 \\
7x_1 + 8x_2 + 9x_3 = 1
\end{cases}
$$

### Example for the system:

$$
\begin{cases}
2x_1 - 3x_2 = 7 \\
3x_1 + 5x_2 = 2
\end{cases}
$$

#### Step 1: Write the coefficient matrix and constants vector

The coefficient matrix \( A \) and the constants vector \( b \) are:

$$
A = \begin{bmatrix}
2 & -3 \\
3 & 5
\end{bmatrix}
,\quad
b = \begin{bmatrix}
7 \\
2
\end{bmatrix}
$$

#### Step 2: Calculate the determinant of the coefficient matrix

The determinant of matrix \( A \) is given by:

$$
\text{det}(A) = 2 \times 5 - (-3) \times 3 = 10 + 9 = 19
$$

#### Step 3: Create matrices for \( x_1 \) and \( x_2 \)

To solve for \( x_1 \), replace the first column of matrix \( A \) with vector \( b \):

$$
A_1 = \begin{bmatrix}
7 & -3 \\
2 & 5
\end{bmatrix}
$$

The determinant of matrix \( A_1 \) is:

$$
\text{det}(A_1) = 7 \times 5 - (-3) \times 2 = 35 + 6 = 41
$$

To solve for \( x_2 \), replace the second column of matrix \( A \) with vector \( b \):

$$
A_2 = \begin{bmatrix}
2 & 7 \\
3 & 2
\end{bmatrix}
$$

The determinant of matrix \( A_2 \) is:

$$
\text{det}(A_2) = 2 \times 2 - 7 \times 3 = 4 - 21 = -17
$$

#### Step 4: Apply Cramer's Rule

Using Cramer's Rule, we solve for \( x_1 \) and \( x_2 \) as follows:

$$
x_1 = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{41}{19}
$$

$$
x_2 = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{-17}{19}
$$

Thus, the solution to the system is:

$$
x_1 = \frac{41}{19}, \quad x_2 = \frac{-17}{19}
$$
