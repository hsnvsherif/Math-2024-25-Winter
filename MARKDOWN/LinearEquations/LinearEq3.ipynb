## Exercises for Students

$$
\left\{\begin{array}{c}{x+2 y-2 z=4} \\ {2 x+y+z=0} \\ {3 x+2 y+z=1}\end{array}\right.\quad
\left\{\begin{array}{c}{x+y+z-t=2} \\ {2 x+y+z=3} \\ {-x+z-t=0} \\ {3 x+2 y-z+2 t=-1}\end{array}\right.\quad
\left\{\begin{array}{c}{x+y-z-t=0} \\ {2 x+3 y-2 z+t=4} \\ {3 x+5 z=0} \\ {-x+y-3 z+2 t=3}\end{array}\right.
$$

$$
\text{1. For the system of equations:}
\left\{
\begin{array}{l}
x + 2y - 2z = 4 \\
2x + y + z = 0 \\
3x + 2y + z = 1
\end{array}
\right.
$$

First, subtract the second equation from the third equation to eliminate \( y \):
$$
(3x + 2y + z) - (2x + y + z) = 1 - 0
$$
Simplifying:
$$
x + y = 1 \quad \Rightarrow \quad y = 1 - x
$$

Substitute \( y = 1 - x \) into the second equation:
$$
2x + (1 - x) + z = 0
$$
Simplifying:
$$
x + z = -1 \quad \Rightarrow \quad z = -1 - x
$$

Now substitute \( y = 1 - x \) and \( z = -1 - x \) into the first equation:
$$
x + 2(1 - x) - 2(-1 - x) = 4
$$
Simplifying:
$$
x + 2 - 2x + 2 + 2x = 4 \quad \Rightarrow \quad x + 4 = 4 \quad \Rightarrow \quad x = 0
$$

Substitute \( x = 0 \) into \( y = 1 - x \) and \( z = -1 - x \):
$$
y = 1, \quad z = -1
$$

Thus, the solution is:
$$
x = 0, \quad y = 1, \quad z = -1
$$

$$
\text{2. For the second system of equations:}
\left\{
\begin{array}{l}
x + y + z - t = 2 \\
2x + y + z = 3 \\
-x + z - t = 0 \\
3x + 2y - z + 2t = -1
\end{array}
\right.
$$

First, subtract the second equation from the first:
$$
(x + y + z - t) - (2x + y + z) = 2 - 3
$$
Simplifying:
$$
-x - t = -1 \quad \Rightarrow \quad x + t = 1 \quad \Rightarrow \quad t = 1 - x
$$

Next, substitute \( t = 1 - x \) into the third equation:
$$
-x + z - (1 - x) = 0
$$
Simplifying:
$$
-x + z - 1 + x = 0 \quad \Rightarrow \quad z = 1
$$

Substitute \( z = 1 \) and \( t = 1 - x \) into the second equation:
$$
2x + y + 1 = 3 \quad \Rightarrow \quad 2x + y = 2 \quad \Rightarrow \quad y = 2 - 2x
$$

Substitute \( z = 1 \), \( t = 1 - x \), and \( y = 2 - 2x \) into the fourth equation:
$$
3x + 2(2 - 2x) - 1 + 2(1 - x) = -1
$$
Simplifying:
$$
3x + 4 - 4x - 1 + 2 - 2x = -1 \quad \Rightarrow \quad -3x + 5 = -1 \quad \Rightarrow \quad -3x = -6 \quad \Rightarrow \quad x = 2
$$

Substitute \( x = 2 \) into \( y = 2 - 2x \), \( z = 1 \), and \( t = 1 - x \):
$$
y = -2, \quad z = 1, \quad t = -1
$$

Thus, the solution is:
$$
x = 2, \quad y = -2, \quad z = 1, \quad t = -1
$$

$$
\text{3. For the third system of equations:}
\left\{
\begin{array}{l}
x + y - z - t = 0 \\
2x + 3y - 2z + t = 4 \\
3x + 5z = 0 \\
-x + y - 3z + 2t = 3
\end{array}
\right.
$$

From the third equation, solve for \( x \):
$$
3x + 5z = 0 \quad \Rightarrow \quad x = -\frac{5}{3}z
$$

Substitute \( x = -\frac{5}{3}z \) into the first equation:
$$
-\frac{5}{3}z + y - z - t = 0
$$
Simplifying:
$$
y - \frac{8}{3}z - t = 0 \quad \Rightarrow \quad y = \frac{8}{3}z + t
$$

Substitute \( x = -\frac{5}{3}z \) and \( y = \frac{8}{3}z + t \) into the second equation:
$$
2\left(-\frac{5}{3}z\right) + 3\left(\frac{8}{3}z + t\right) - 2z + t = 4
$$
Simplifying:
$$
-\frac{10}{3}z + 8z + 3t - 2z + t = 4
$$
Combining terms:
$$
\frac{14}{3}z + 4t = 4 \quad \Rightarrow \quad 14z + 12t = 12
$$
Simplifying:
$$
7z + 6t = 6
$$

Now substitute \( x = -\frac{5}{3}z \) and \( y = \frac{8}{3}z + t \) into the fourth equation:
$$
-\left(-\frac{5}{3}z\right) + \left(\frac{8}{3}z + t\right) - 3z + 2t = 3
$$
Simplifying:
$$
\frac{5}{3}z + \frac{8}{3}z + t - 3z + 2t = 3 \quad \Rightarrow \quad \frac{13}{3}z + 3t = 3
$$
Multiplying through by 3:
$$
13z + 9t = 9
$$

Solve the system:
$$
7z + 6t = 6 \quad \text{(1)}
$$
$$
13z + 9t = 9 \quad \text{(2)}
$$

Multiply equation (1) by 3:
$$
21z + 18t = 18 \quad \text{(3)}
$$
Multiply equation (2) by 2:
$$
26z + 18t = 18 \quad \text{(4)}
$$
Subtract equation (3) from equation (4):
$$
5z = 0 \quad \Rightarrow \quad z = 0
$$

Substitute \( z = 0 \) into equations (1) and (2):
$$
7(0) + 6t = 6 \quad \Rightarrow \quad 6t = 6 \quad \Rightarrow \quad t = 1
$$

Substitute \( z = 0 \) and \( t = 1 \) into the expressions for \( x \) and \( y \):
$$
x = -\frac{5}{3}(0) = 0, \quad y = \frac{8}{3}(0) + 1 = 1
$$

Thus, the solution is:
$$
x = 0, \quad y = 1, \quad z = 0, \quad t = 1
$$
