## Exercises for Students

1. Solve the system of linear equations using the inverse matrix method:

$$
\begin{cases}
x + 2y + 3z = 5, \\
2y + 3z = 4, \\
3z = 3.
\end{cases}
$$

2. Solve the system of linear equations using the inverse matrix method:

$$
\begin{cases}
x_1 + 2x_2 + 3x_3 = 41, \\
4x_1 + 5x_2 + 6x_3 = 93, \\
7x_1 + 8x_2 + 9x_3 = 145.
\end{cases}
$$

$$
\text{1. Solve the system using the inverse matrix method:}
\begin{cases}
x + 2y + 3z = 5, \\
2y + 3z = 4, \\
3z = 3.
\end{cases}
$$

First, write the system in matrix form \(A \mathbf{x} = \mathbf{b}\):
$$
\begin{bmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
5 \\
4 \\
3
\end{bmatrix}
$$

Now, find the inverse of matrix \(A\). The inverse of a triangular matrix is straightforward to compute, and the inverse matrix \(A^{-1}\) is:
$$
A^{-1} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & -\frac{1}{6} \\
0 & 0 & \frac{1}{3}
\end{bmatrix}
$$

Now, solve for \( \mathbf{x} \) by multiplying both sides of the matrix equation by \( A^{-1} \):
$$
\mathbf{x} = A^{-1} \mathbf{b}
$$

Perform the multiplication:
$$
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & -\frac{1}{6} \\
0 & 0 & \frac{1}{3}
\end{bmatrix}
\begin{bmatrix}
5 \\
4 \\
3
\end{bmatrix}
$$

First row:
$$
x = 1 \times 5 = 5
$$

Second row:
$$
y = \frac{1}{2} \times 4 - \frac{1}{6} \times 3 = 2 - 0.5 = 1.5
$$

Third row:
$$
z = \frac{1}{3} \times 3 = 1
$$

Thus, the solution is:
$$
x = 5, \quad y = 1.5, \quad z = 1
$$

$$
\text{2. Solve the system using the inverse matrix method:}
\begin{cases}
x_1 + 2x_2 + 3x_3 = 41, \\
4x_1 + 5x_2 + 6x_3 = 93, \\
7x_1 + 8x_2 + 9x_3 = 145.
\end{cases}
$$

Write the system in matrix form \(A \mathbf{x} = \mathbf{b}\):
$$
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
x_3
\end{bmatrix}
=
\begin{bmatrix}
41 \\
93 \\
145
\end{bmatrix}
$$

The determinant of matrix \(A\) is \(0\), so matrix \(A\) is singular and does not have an inverse. This means the system of equations has no unique solution (either no solution or infinitely many solutions).
