## Finding the Equation of a Line

1. **Formula for the Slope**:  
   The slope \( m \) of a line passing through two points \( (x_1, y_1) \) and \( (x_2, y_2) \) is given by:
   $$
   m = \frac{y_2 - y_1}{x_2 - x_1}
   $$

2. **Substitute the Given Points**:  
   Here, \( A(1,2) \) and \( B(3,4) \), so:
   $$
   m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
   $$

3. **Use the Point-Slope Form**:  
   The equation of a line in **point-slope form** is:
   $$
   y - y_1 = m(x - x_1)
   $$
   Using point \( A(1,2) \):
   $$
   y - 2 = 1(x - 1)
   $$

4. **Convert to Slope-Intercept Form**:  
   $$
   y - 2 = x - 1
   $$
   $$
   y = x + 1
   $$

**Final Answer**:  
The equation of the line is:
$$
y = x + 1
$$

## Finding the Equation of a Parallel Line

1. **Understanding Parallel Lines**:  
   Two lines are **parallel** if they have the **same slope**.  
   The given line equation is:
   $$
   y = 2x + 3
   $$
   Here, the slope is \( m = 2 \).

2. **Use the Point-Slope Form**:  
   The equation of a line in **point-slope form** is:
   $$
   y - y_1 = m(x - x_1)
   $$
   Given that the line passes through \( A(1,2) \), we substitute:
   $$
   y - 2 = 2(x - 1)
   $$

3. **Convert to Slope-Intercept Form**:  
   $$
   y - 2 = 2x - 2
   $$
   $$
   y = 2x
   $$

**Final Answer**:  
The equation of the required parallel line is:
$$
y = 2x

## Finding the Equation of a Perpendicular Line

1. **Understanding Perpendicular Lines**:  
   Two lines are **perpendicular** if the product of their slopes is \( -1 \).  
   The given line equation is:
   $$
   y = 2x + 3
   $$
   Here, the slope is \( m_1 = 2 \).

   The slope \( m_2 \) of the line perpendicular to this one satisfies:
   $$
   m_1 \times m_2 = -1
   $$  
   Substituting \( m_1 = 2 \), we get:
   $$
   2 \times m_2 = -1 \quad \Rightarrow \quad m_2 = -\frac{1}{2}
   $$

2. **Use the Point-Slope Form**:  
   The equation of a line in **point-slope form** is:
   $$
   y - y_1 = m(x - x_1)
   $$
   Given that the line passes through \( A(1,2) \) and has slope \( m_2 = -\frac{1}{2} \), we substitute:
   $$
   y - 2 = -\frac{1}{2}(x - 1)
   $$

3. **Convert to Slope-Intercept Form**:  
   $$
   y - 2 = -\frac{1}{2}x + \frac{1}{2}
   $$  
   $$
   y = -\frac{1}{2}x + \frac{5}{2}
   $$

**Final Answer**:  
The equation of the required perpendicular line is:
$$
y = -\frac{1}{2}x + \frac{5}{2}
$$

1. **Equation of the Line**:

   The given equation of the line is:
   $$
   y = 2x + 3
   $$

2. **Convert the Line to Standard Form**:

   The general form of the line equation is \( Ax + By + C = 0 \). To convert the equation \( y = 2x + 3 \) into this form, we rearrange it:
   $$
   y - 2x - 3 = 0
   $$
   So, the line is in the form \( Ax + By + C = 0 \), where \( A = -2 \), \( B = 1 \), and \( C = -3 \).

3. **Formula for the Distance from a Point to a Line**:

   The formula for the distance \( d \) from a point \( (x_1, y_1) \) to a line \( Ax + By + C = 0 \) is given by:
   $$
   d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
   $$

4. **Substitute the Values**:

   For point \( A(1, 2) \), we have \( x_1 = 1 \) and \( y_1 = 2 \). Substituting the values of \( A = -2 \), \( B = 1 \), and \( C = -3 \) into the formula:
   $$
   d = \frac{|(-2)(1) + (1)(2) - 3|}{\sqrt{(-2)^2 + (1)^2}}
   $$

5. **Simplify**:

   First, calculate the numerator:
   $$
   (-2)(1) + (1)(2) - 3 = -2 + 2 - 3 = -3
   $$

   Now, calculate the denominator:
   $$
   \sqrt{(-2)^2 + (1)^2} = \sqrt{4 + 1} = \sqrt{5}
   $$

   So, the distance is:
   $$
   d = \frac{|-3|}{\sqrt{5}} = \frac{3}{\sqrt{5}} = \frac{3\sqrt{5}}{5}
   $$

Thus, the distance from point \( A(1, 2) \) to the line \( y = 2x + 3 \) is:
$$
d = \frac{3\sqrt{5}}{5}
$$

### Two-Point Formula for the Equation of a Line

Given that the line intersects the coordinate axes at points \( A(2, 0) \) and \( B(0, 3) \), we can use the **two-point formula** to find the equation of the line.

1. **Two-Point Formula**:
   The equation of a line passing through points \( (x_1, y_1) \) and \( (x_2, y_2) \) is:
   $$
   y - y_1 = \frac{y_2 - y_1}{x_2 - x_1} (x - x_1)
   $$

2. **Substitute the Points \( A(2, 0) \) and \( B(0, 3) \)**:
   Substituting \( x_1 = 2, y_1 = 0, x_2 = 0, y_2 = 3 \), we get:
   $$
   y - 0 = \frac{3 - 0}{0 - 2} (x - 2)
   $$

3. **Simplify the Equation**:
   $$
   y = -\frac{3}{2} (x - 2)
   $$

4. **Final Equation**:
   Expanding and simplifying further:
   $$
   y = -\frac{3}{2}x + 3
   $$

Thus, the equation of the line is:
$$
y = -\frac{3}{2}x + 3
$$

### Angle Between a Line and the \( x \)-Axis

We are given the line equation:

$$
y = x + 3
$$

1. **Slope of the Line**:
   The slope of the line \( m \) is the coefficient of \( x \), which is 1. So, we have:

$$
m = 1
$$

2. **Formula for the Angle**:
   The angle \( \theta \) between the line and the \( x \)-axis can be calculated using the formula:

$$
\tan(\theta) = |m|
$$

Therefore:

$$
\theta = \tan^{-1}(m) = \tan^{-1}(1)
$$

3. **Calculate the Angle**:
   Using the fact that \( \tan^{-1}(1) = 45^\circ \), we get:

$$
\theta = 45^\circ
$$

Thus, the angle between the line \( y = x + 3 \) and the \( x \)-axis is:

$$
\theta = 45^\circ
$$

### Perpendicular Vector to the Line \( x + y + 1 = 0 \)

1. **Line Equation**: 
   The line equation is:
   $$
   x + y + 1 = 0
   $$

2. **Direction Vector of the Line**: 
   From the line equation \( x + y + 1 = 0 \), we can extract the direction vector \( \mathbf{v} = \begin{bmatrix} 1 \\ 1 \end{bmatrix} \).

3. **Perpendicular Vector**: 
   The vector perpendicular to the line will be the vector that is orthogonal to \( \mathbf{v} \). This can be obtained by swapping the components of the direction vector and changing the sign of one of the components. Thus, the perpendicular vector is:
   $$
   \mathbf{v_\perp} = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
   $$

Therefore, a vector perpendicular to the line \( x + y + 1 = 0 \) is:
$$
\mathbf{v_\perp} = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
$$
