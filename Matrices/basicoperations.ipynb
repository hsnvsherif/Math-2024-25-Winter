## Exercises for Students

Perform the following matrix operations:

### 1. Sum of Matrices
Given matrices \( A_1 \) and \( B_1 \):

\[
A_1 = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad B_1 = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
\]

Calculate the sum \( A_1 + B_1 \).

### 2. Scalar Multiplication
Given matrix \( A_2 \) and scalar \( k \):

\[
A_2 = \begin{bmatrix} 4 & 2 \\ 12 & 4 \end{bmatrix}, \quad k = \frac{1}{4}
\]

Calculate the scalar multiplication \( k \cdot A_2 \).

### 3. Matrix Products
Given matrices \( A_3 \) and \( B_3 \):

\[
A_3 = \begin{bmatrix} 3 & 2 \\ 2 & 4 \end{bmatrix}, \quad B_3 = \begin{bmatrix} 4 & 1 \\ 6 & 4 \end{bmatrix}
\]

Calculate the products \( A_3 \cdot B_3 \) and \( B_3 \cdot A_3 \).


```python
import numpy as np

# Define matrices
A1 = np.array([[1, 2], [3, 4]])
B1 = np.array([[5, 6], [7, 8]])

A2 = np.array([[4, 2], [12, 4]])
k = 1 / 4

A3 = np.array([[3, 2], [2, 4]])
B3 = np.array([[4, 1], [6, 4]])

# Perform operations
sum_result = A1 + B1
scalar_result = k * A2
product_A_B = np.dot(A3, B3)
product_B_A = np.dot(B3, A3)

# Convert matrix to LaTeX format
def to_latex(matrix):
    rows = [" & ".join(f"{value:.2f}".rstrip('0').rstrip('.') for value in row) for row in matrix]
    return "\\begin{bmatrix}\n" + " \\\\\n".join(rows) + "\n\\end{bmatrix}"

# Print LaTeX output
print("Exercise 1: Sum of Matrices")
print(f"$A + B = {to_latex(sum_result)}$\n")

print("Exercise 2: Scalar Multiplication")
print(f"$k \\cdot A = {to_latex(scalar_result)}$\n")

print("Exercise 3: Matrix Products")
print(f"$A \\cdot B = {to_latex(product_A_B)}$\n")
print(f"$B \\cdot A = {to_latex(product_B_A)}$\n")
```
